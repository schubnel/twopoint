```python
import matplotlib.pyplot as plt
from math import pi, cos, sin
from random import random,seed
seed(999999)
def point(h, k, r):
    theta = random() * 2 * pi
    return h + cos(theta) * r, k + sin(theta) * r
n=2 # n=1 or 4. (same number of points!)
circle_r=5.
n_circles=n*25*2 #25
c_points=int(60/n)
ntotal=n_circles*c_points
points=[]
d_circle=10.
for m in range(0,n_circles):
    circle_x=random()*d_circle*4 - d_circle*2
    circle_y=random()*d_circle*4 - d_circle*2
    p_xy = [point(circle_x,circle_y,circle_r) for _ in range(c_points)]
    points=points+p_xy
bound=8.   
D=points
Dsel=[x for x in D if (abs(x[0])<bound and abs(x[1])<bound )]
l=list(zip(*Dsel))
plt.plot(l[0],l[1],'.',color='g')
plt.grid(color='k', linestyle=':', linewidth=1)
plt.axes().set_aspect('equal')#, 'datalim')

plt.xlim([-bound,bound])
plt.ylim([-bound,bound])
plt.show()
#print(D)
Dsel=[x for x in D if (abs(x[0])<bound and abs(x[1])<bound )]
print(len(Dsel))
```

```python
w=5
num=int(len(Dsel))
num=num*w
def rpoint():
    x=random()*bound*2-bound
    y=random()*bound*2-bound
    return x,y
R=[rpoint() for e in range(0,num)]
l=list(zip(*R))
print(len(R))
plt.plot(l[0],l[1],'.',color='b')
plt.grid(color='k', linestyle=':', linewidth=1)
plt.axes().set_aspect('equal')#, 'datalim')

plt.xlim([-bound,bound])
plt.ylim([-bound,bound])
plt.show()
```

```python

import numpy as np
import itertools
def linalg_norm(a,b):
    return np.linalg.norm(a-b, axis=1)

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(18,4))  # 1 row, 3 columns
# DD
p=itertools.combinations(Dsel,2)
#a=list(p)
#print(a[0:3])
a,b=zip(*p)
DDdst=linalg_norm(np.array(a),np.array(b))
nDD=len(DDdst)
print('length DD: ',nDD)
DDn,bin,patches=ax1.hist(DDdst, bins=np.arange(0,20.2,.2), normed=1, facecolor='green', alpha=0.55)
ax1.set_title('DD')
# RR
p=itertools.combinations(R,2)
a,b=zip(*p)
RRdst=linalg_norm(np.array(a),np.array(b))
nRR=len(RRdst)
print('length RR: ',nRR)
RRn,bin,patches=ax2.hist(RRdst, bins=np.arange(0,20.2,.2), normed=1, facecolor='blue', alpha=0.55)
ax2.set_title('RR')
# DR
DR=np.outer(R,Dsel).ravel()
nDR=len(DR)
print('length DR: ',nDR)
DRn,bin,patches=ax3.hist(DR, bins=np.arange(0,20.2,.2), normed=1, facecolor='gold', alpha=0.55)
ax3.set_title('DR')
#plt.tight_layout() 
plt.show()
```

```python
%%latex
Now we measure the two-point correlation function by using the estimator by Landy & Szlazy (1993):}
\begin{align}
    
\eta = \frac{DD-2DR+RR}{RR}
\end{align}
where DD, RR and DR are the normalized numbers of pairs in a given bin
```

```python

eta=(DDn/nDD)-2*DRn/nDR+RRn/nRR
eta=eta*nRR/RRn
eta=eta/25.
x=np.arange(0,100)*.2+.1
plt.plot(x,eta,color='purple',alpha=0.7)
plt.xlim([6.,12.])
plt.ylim([.8,1.6])
plt.title('Two-Point Correlation Function Estimator')
plt.xlabel('distance')
plt.show()
```
